```python
#importing packages & libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
#changing configuration of plots
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
pd.options.mode.chained_assignment = None
```

```python
#importing data
df = pd.read_csv(r'C:\Users\cambw\Downloads\movies.csv')
```

```python
#see if data loaded correctly
df
```

```python
#checking for missing values
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#looking at data types for variables
print(df.dtypes)
```

```python
#exploring distribution of data
df.boxplot(column=['gross'])
```

```python
#dropping duplicates just in case
df.drop_duplicates()
```

```python
#exploring top grossing movies
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#not surprised to see Avatar & Avengers up there
#wanna explore relationship between movie gross earnings & budget
#assume they are correlated due to big blockbluster nature of hollywood but might be able to isolate flops too
#double checking my assumption
sns.regplot(x="gross", y="budget", data=df)
```

```python
#linear coorelation b/w gross earnings & budget as expected 
#are gross earnings & IMDB scores correlated?
sns.regplot(x="score", y="gross", data=df)
```

```python
#not really but somewhat linear correlation
#exploring what else is correlated to what & to what degree through some statistical tests
#Pearson Method
df.corr(method ='pearson')
```

```python
#Kendall Method
df.corr(method ='kendall')
```

```python
#Spearman Method
df.corr(method ='spearman')
```

```python
#interesting to see values change depending on methods used
#lets explore the data not from statistical perspective but from visual perspective
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
#gonna add the categorical stuff in the heatmap matrix too
#gonna use random stuff/monte carlo experiment to determine values
#using factorize to assign a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
#okay, above stuff is statistical information
#let's take a look at it visually now like I did before
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
#name & company are correlated?
#i guess it makes sense since many companies have many movies with same IP
#director & gross correlated? 
#makes sense, ppl go to theatres to see the masters' work
#further exploration of correlations
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()
corr_pairs = correlation_mat.unstack()
print(corr_pairs)
```

```python
#sorting b/w pairs of correlations to visually see a bit better
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
print(sorted_pairs)
```

```python
#now that I have things organized, wanna see just the pairs that have a strong to moderate correlation
#kinda discretionary, but I will choose .5 as base value
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]
print(strong_pairs)
```

```python
#makes sense
#companies will recast same star for IP franchises
#writers will typically be hired within same company for IP congruency across films
#directors hired by same company for similar reasons
#wanna transition out of correlation analysis now
#wanna focus on top companies all time by gross revneue
#suspect to see giants like Warner Bros. & Disney at the top, but you never know
CompanyGrossSum = df.groupby('company')[["gross"]].sum()
CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]
CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 
CompanyGrossSumSorted
```

```python
#as expected; however, surprised to see New Line Cinema, Touchstone Pictures?
#never heard of them
#gonna do some research to cross-check
#ok, so New Line has Lord of the Rings, so that makes sense
#touchstone has some good properties too
#never heard of touchstone's highest grossing film armeggedon, but have watched other popular stuff like Knighmare B4 Christmas
#gonna take a look at top companies by gross by year
CompanyGrossSum = df.groupby(['company', 'year'])[["gross"]].sum()
CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company','year'], ascending = False)[:15]
CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 
CompanyGrossSumSorted
```

```python
#damn, as expected, Disney crushed it with recent Star Wars stuff
#same with Marvel franchises 
#im curious about Universal Pictures tho in 2015
#going to cross-check
#yeah, that was the year when Jurrasic Park remake came out, which was massive
#i wanna revisit gross vs. budget relationship
plt.scatter(x=df['budget'], y=df['gross'], alpha=0.5)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```
